```python
from sklearn.manifold import TSNE
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Basic plot conf
plt.rcParams['legend.title_fontsize'] = 'x-large'
plt.rcParams['legend.fontsize'] = 'x-large'
plt.rcParams['figure.figsize'] = (20,15)
```

```python
# Load the Grain dataset
grain = pd.read_csv('../input/seeds.csv', sep=',')
# grainV2 = pd.read_excel('../input/fgyearbooktablesfull.xls')
```

```python
# Intialize t-SNE with a learning_rate = 150
model = TSNE(learning_rate=150)
grain_transformed = model.fit_transform(grain)
print(f'Number of t-SNE features: {grain_transformed.shape[1]}')
xs = grain_transformed[:,0]
ys = grain_transformed[:,1]
```

```python
# PLot the resulting features according
sns.scatterplot(xs, ys, hue=grain['1'])
plt.grid()
```
